```python
lengths = np.arange(15,51)
files = ["pulsesWOinterferometer/C3pulse shape with delay line {}.7 mm00000.txt".format(int(l)) for l in lengths]
```

```python
fig, axes = plt.subplots(6, 6, figsize=(10, 10), dpi=120)
for ax, file, length in zip(np.ravel(axes), files, lengths):
    data = np.loadtxt(file, skiprows=5)
    ax.plot(data[:,0], data[:,1])
    ax.set_title("DL = {}.7mm".format(int(length)))
plt.tight_layout()
```
